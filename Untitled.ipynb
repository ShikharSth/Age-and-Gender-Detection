```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
from PIL import Image, ImageOps
from sklearn.model_selection import train_test_split

from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense
from keras import optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import image_dataset_from_directory
```

```python
images = []
ages = []
genders = []
directory_path = 'dataset/archive-2/UTKFace/'

for i in os.listdir(directory_path)[0:8000]:
    split = i.split('_')
    ages.append(int(split[0]))
    genders.append(int(split[1]))
    with Image.open(directory_path + i) as img:
        images.append(img.copy())
```

```python
images = pd.Series(list(images), name = 'Images')
ages = pd.Series(list(ages), name = 'Ages')
genders = pd.Series(list(genders), name = 'Genders')

df = pd.concat([images, ages, genders], axis=1)
df
```

```python
df = df.reset_index(drop=True)

x = []
y = []

for i in range(len(df)):
    img = df.loc[i, 'Images'].resize((200, 200), Image.LANCZOS)  # Resize image
    ar = np.asarray(img)  
    x.append(ar)
    agegen = [int(df.loc[i, 'Ages']), int(df.loc[i, 'Genders'])]
    y.append(agegen)
x = np.array(x)

y_age = df['Ages']
y_gender = df['Genders']
```

```python
min_samples = 2
valid_classes_age = df['Ages'].value_counts()[df['Ages'].value_counts() >= min_samples].index
valid_classes_gender = df['Genders'].value_counts()[df['Genders'].value_counts() >= min_samples].index

df_filtered = df[df['Ages'].isin(valid_classes_age) & df['Genders'].isin(valid_classes_gender)]

# Reset index after filtering
df_filtered = df_filtered.reset_index(drop=True)

# Prepare filtered data
x_filtered = []
y_filtered = []

for i in range(len(df_filtered)):
    img = df_filtered.loc[i, 'Images'].resize((200, 200), Image.LANCZOS)
    ar = np.asarray(img)
    x_filtered.append(ar)
    agegen = [int(df_filtered.loc[i, 'Ages']), int(df_filtered.loc[i, 'Genders'])]
    y_filtered.append(agegen)
x_filtered = np.array(x_filtered)

y_age_filtered = df_filtered['Ages']
y_gender_filtered = df_filtered['Genders']

# Step 3: Split the data
x_train_age, x_test_age, y_train_age, y_test_age = train_test_split(x_filtered, y_age_filtered, test_size=0.2, stratify=y_age_filtered)
x_train_gender, x_test_gender, y_train_gender, y_test_gender = train_test_split(x_filtered, y_gender_filtered, test_size=0.2, stratify=y_gender_filtered)
```

```python
# Age Model
agemodel = Sequential()
agemodel.add(Conv2D(32, (3,3), activation='relu', input_shape=(200, 200, 3)))
agemodel.add(MaxPooling2D((2,2)))
agemodel.add(Conv2D(64, (3,3), activation='relu'))
agemodel.add(MaxPooling2D((2,2)))
agemodel.add(Conv2D(128, (3,3), activation='relu'))
agemodel.add(MaxPooling2D((2,2)))
agemodel.add(Flatten())
agemodel.add(Dense(64, activation='relu'))
agemodel.add(Dropout(0.5))
agemodel.add(Dense(1, activation='relu'))

agemodel.compile(loss='mean_squared_error',
                 optimizer=optimizers.Adam(learning_rate=0.0001))

# Gender Model
genmodel = Sequential()
genmodel.add(Conv2D(32, (3,3), activation='relu', input_shape=(200, 200, 3)))
genmodel.add(MaxPooling2D((2,2)))
genmodel.add(Conv2D(64, (3,3), activation='relu'))
genmodel.add(MaxPooling2D((2,2)))
genmodel.add(Conv2D(128, (3,3), activation='relu'))
genmodel.add(MaxPooling2D((2,2)))
genmodel.add(Flatten())
genmodel.add(Dense(64, activation='relu'))
genmodel.add(Dropout(0.5))
genmodel.add(Dense(1, activation='sigmoid'))

genmodel.compile(loss='binary_crossentropy',
                 optimizer=optimizers.Adam(learning_rate=0.0001),
                 metrics=['accuracy'])
```

```python
datagen = ImageDataGenerator(
      rescale=1./255., width_shift_range = 0.1, height_shift_range = 0.1, horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale=1./255)

train1 = datagen.flow(x_train_age, y_train_age, batch_size=32)

test1 = test_datagen.flow(
        x_test_age, y_test_age,
        batch_size=32)

history1 = agemodel.fit(train1, epochs=30, shuffle=True, validation_data=test1)
```

```python
datagen = ImageDataGenerator(
      rescale=1./255., width_shift_range = 0.1, height_shift_range = 0.1, horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale=1./255)

train2 = datagen.flow(x_train_gender, y_train_gender, batch_size=32)

test2 = test_datagen.flow(
        x_test_gender, y_test_gender,
        batch_size=32)

history2 = genmodel.fit(train2, epochs=30, shuffle=True, validation_data=test2)
```
